```python
import os
import sys
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
folder_path  = Path.joinpath(Path(os.getcwd()), 'csv')
```

```python
headers = ['href', 'title', 'full_text', 'phone', 'region', 'city', 'real_estate', 'type', 'marketplace']
```

```python
dfs=[]
for the_file in os.listdir(folder_path):
    file_path = os.path.join(folder_path, the_file)
    if os.path.isfile(file_path) and  the_file.endswith('.csv'):
        print('working with file: ',the_file)
        df = pd.read_csv(file_path, sep=',', encoding='utf8')
        for head in headers:
            if head not in df.columns:
                df[head]=[None]*len(df)
        df=df[df.href!='href']
        df['region'] = the_file.split('-')[2]
        if not df.href.isnull().sum():
            df['city'] = df['href'].apply(lambda x: x.split('/')[3])
            df['marketplace'] = df['href'].apply(lambda x: x.split('/')[2])
        else:
            df['city']        = 'moskva'
            df['marketplace'] = 'm.avito.ru'
            df.href.fillna('-',inplace=True)
        df['real_estate'] = the_file.split('-')[3]
        df['type'] = the_file.split('-')[-1].split('.')[0]
        dfs.append(df)
```

```python
for dfi in dfs[1:]:
    dfs[0] = pd.concat([dfs[0],dfi], ignore_index=True, sort=False)
```

```python
df = dfs[0]
df = df[headers]
```
